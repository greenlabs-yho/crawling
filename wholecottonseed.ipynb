```python
import pytz
from datetime import datetime
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd
from requests.adapters import HTTPAdapter, Retry

soup = BeautifulSoup(html, 'html.parser')
items = soup.find_all('option')
```

```python
def get_data(p):
    # 추출된 데이터를 저장할 딕셔너리
    data = {}

    if p.find('strong'):
        # 회사명 추출
        data['Company Name'] = p.find('strong').text

        # 나머지 정보 추출
        for br in p.find_all('br'):
            next_s = br.nextSibling
            if next_s and isinstance(next_s, str):
                if 'Ph:' in next_s:
                    data['Phone'] = next_s.replace('Ph:', '').strip()
                elif 'Email:' in next_s:
                    data['Email'] = next_s.next_sibling.text
                elif 'Website:' in next_s:
                    data['Website'] = next_s.next_sibling.text
                else:
                    data['address'] = next_s.replace('\t', '')
            
    return data
```

```python
from selenium import webdriver

# 웹드라이버 설정 (예: Chrome의 경우)
driver = webdriver.Chrome()
```

```python
import time
result_list = []
for item in items:
    url = item['value']
    print(url)
    driver.get(url)
    time.sleep(3)
    # 페이지의 HTML 가져오기
    html_content = driver.page_source
    soup = BeautifulSoup(html_content, 'html.parser')

    
    parent = soup.find('div', class_="default-txt")

    p_list = parent.find_all('p')
    #print(p_list)
    for i, p in enumerate(p_list):
        if i == 0:
            continue
        result = get_data(p)
        if result:
            result_list.append(result)
```

```python
# 웹드라이버 종료
driver.quit()
```

```python
import pandas as pd
df = pd.DataFrame(result_list)
df.to_csv("./output/wholecottonseed_20240119_1445.csv", index=False)
```
