```python
j_data = response.json()
soup = BeautifulSoup(j_data["html_content"], 'html.parser')
```

```python
div_list = soup.find_all('div', class_='exp-arrow')
```

```python
div_list
```

```python

result_list = []
for div in div_list:
    item = {
        "name": div.find('h3').text.strip(),
        "link": div.find('a').get('href'),
        "address": div.find("p").text.strip().replace('\xa0', ' '),
    }
    mail = div.find("li", class_="teampress-email")
    if mail:
        item["mail"] = mail.find("a").get('href').replace('mailto:', '')
    mobile = div.find("li", class_="teampress-mobile")
    if mobile:
        item["mobile"] = mobile.find("a").get('href').replace('tel:', '')
    
    web = div.find('li', class_="tp-ctsocial teampress-Website")
    if web:
        item["web"] = web.find("a").get('href')
    
    if item['link'].startswith('https://michiganbean.com'):
        sub_response = session.get(item['link'], headers=headers)
        sub_soup = BeautifulSoup(sub_response.text, 'html.parser')
        category = sub_soup.find('div', class_='mb-meta team-cat-info')
        item['SKU'] = category.text.replace('Categories:', '').strip().strip()

        
    print(item)
    result_list.append(item)
```

```python
df = pd.DataFrame(result_list)
df.to_csv('./output/michiganbean_20240213.csv', index=False, quoting=1)
```
